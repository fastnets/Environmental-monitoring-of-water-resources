## Объем сточных вод

```python
import pandas as pd
df = pd.read_excel('Pril-Region_Pokaz_2024/Раздел 7 - Зем ресурсы и охр окр прир среды.xlsx', sheet_name=9, skiprows=5, skipfooter=3)
df.rename(columns={'Unnamed: 0':'Регион', '20142)':'2014'}, inplace=True)
df['Регион'] = df['Регион'].apply(lambda x: x.strip())
df
```

```python
df = df.melt(id_vars=['Регион'], var_name='Год', value_name='Объем_сточных_вод_млн_м3')
df = df[['Регион', 'Год', 'Объем_сточных_вод_млн_м3']]
```

```python
df
```

## Инвестиции

Инвестиции 1991-2008

```python
data1 = pd.read_excel("data (1).xls", skiprows=2)
data1.drop(0, inplace=True)
data1.rename(columns={'Unnamed: 0':'Регион', 'Unnamed: 1':'Вид оос'}, inplace=True)
data1['Регион'] = data1['Регион'].apply(lambda x: x.strip())
data1 = pd.DataFrame(data1.drop('Вид оос', axis=1).groupby(['Регион']).sum())
data1.reset_index(inplace=True)
data1
```

Инвестиции 2009-2016

```python
data2 = pd.read_excel("data (2).xls", skiprows=2)
data2.drop(0, inplace=True)
data2.drop(['Unnamed: 0', 'Unnamed: 2'], axis=1, inplace=True)
data2.rename(columns={'Unnamed: 1':'Регион', 'Unnamed: 3':'Вид оос'}, inplace=True)
data2['Регион'] = data2['Регион'].apply(lambda x: x.strip())
data2 = pd.DataFrame(data2.drop('Вид оос', axis=1).groupby(['Регион']).sum())
data2.reset_index(inplace=True)
data2
```

Инвестиции 2017-2024

```python
data3 = pd.read_excel("data (3).xls", skiprows=2)
data3.drop(0, inplace=True)
data3.rename(columns={'Unnamed: 0':'Регион', 'Unnamed: 1':'Вид оос'}, inplace=True)
data3['Регион'] = data3['Регион'].apply(lambda x: x.strip())
data3 = pd.DataFrame(data3.drop('Вид оос', axis=1).groupby(['Регион']).sum())
data3.reset_index(inplace=True)
data3
```

```python
all_invest = pd.merge(data1, data2, how='outer', on='Регион')
all_invest = pd.merge(all_invest, data3, how='outer', on='Регион')
```

```python
all_invest
```

```python
all_invest.info()
```

## Врп

```python
vrp = pd.read_excel('Pril-Region_Pokaz_2024/Раздел 8 - Валовой региональный продукт.xlsx', sheet_name=1, skiprows=8)
vrp.rename(columns={'Unnamed: 0':'Регион'}, inplace=True)
vrp['Регион'] = vrp['Регион'].apply(lambda x: x.strip())
vrp.drop(['Unnamed: 24', 'Unnamed: 25'], axis=1, inplace=True)
vrp
```

```python
vrp_2023 = pd.read_excel("data (4).xls", skiprows=2)
vrp_2023.drop(0, inplace=True)
vrp_2023.drop('Unnamed: 1', axis=1, inplace=True)
vrp_2023.rename(columns={'Unnamed: 0':'Регион'}, inplace=True)
vrp_2023['Регион'] = vrp_2023['Регион'].apply(lambda x: x.strip())
vrp_2023
```

```python
all_vrp = pd.merge(vrp, vrp_2023, how='outer')
```

```python
set(vrp['Регион']) - set(vrp_2023['Регион'])
```

# Объединение датасета

```python
all_invest = all_invest.melt(id_vars=['Регион'], var_name='Год', value_name='Инвестиции_в_ООС')
all_invest = all_invest[['Регион', 'Год', 'Инвестиции_в_ООС']]
```

```python
all_vrp = all_vrp.melt(id_vars=['Регион'], var_name='Год', value_name='ВРП')
all_vrp = all_vrp[['Регион', 'Год', 'ВРП']]
```

```python
all_df = pd.merge(pd.merge(df, all_invest, how='outer', on=['Регион', 'Год']), all_vrp, how='outer', on=['Регион', 'Год'])
```
